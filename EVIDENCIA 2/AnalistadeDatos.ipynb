```python
import pandas as pd

# Carga el dataset
df = pd.read_csv('/content/Hoja2.csv', encoding='latin-1')

# Selecciona solo las columnas numéricas
num_cols = df.select_dtypes(include=['number'])

# Calcula la cantidad de valores válidos sobre el total
valid_values = num_cols.notnull().sum()
total_values = num_cols.shape[0]
valid_percentage = (valid_values / total_values) * 100

# Calcula las estadísticas claves
stats = num_cols.describe()

# Detecta los valores mal registrados o erróneos
outliers = ((num_cols < num_cols.quantile(0.05)) | (num_cols > num_cols.quantile(0.95))).sum()

# Cuenta los valores ausentes
missing_values = num_cols.isnull().sum()

# Resultados
print(" Estadísticas generales de columnas numéricas:\n", stats)
print("\n Cantidad de valores válidos sobre el total:\n", valid_values)
print("\n Porcentaje de valores válidos:\n", valid_percentage)
print("\n Valores ausentes en columnas numéricas:\n", missing_values)
print("\n Valores mal registrados o erróneos (posibles atípicos):\n", outliers)
```

```python
import pandas as pd

# Carga el dataset
df = pd.read_csv('/content/Hoja2.csv', encoding='latin-1')

# Selecciona solo las columnas categóricas
cat_cols = df.select_dtypes(include=['object'])

# Cuenta los registros válidos por columna
valid_counts = cat_cols.notnull().sum()

# Cuenta los registros ausentes por columna
missing_counts = cat_cols.isnull().sum()

# Cuenta la frecuencia de cada categoría dentro de cada columna
category_counts = {col: cat_cols[col].value_counts() for col in cat_cols.columns}

# Resultados
print(" Cantidad de registros válidos en cada columna categórica:\n", valid_counts)
print("\n Valores ausentes en columnas categóricas:\n", missing_counts)
print("\n Frecuencia de cada categoría en cada columna:")
for col, count in category_counts.items():
    print(f"\nCategorías en '{col}':\n", count)
```

```python
#3.Analisis de variables adicionales
import pandas as pd

# Carga el dataset
df = pd.read_csv('/content/Hoja2.csv', encoding='latin-1')

# Selecciona las columnas que no sean completamente numéricas ni categóricas
other_cols = df.select_dtypes(exclude=['number', 'object'])

# Cuenta los registros válidos
valid_counts = other_cols.notnull().sum()

# Cuenta los registros ausentes
missing_counts = other_cols.isnull().sum()

# Verifica los posibles valores mal registrados
outliers = ((other_cols < other_cols.quantile(0.05)) | (other_cols > other_cols.quantile(0.95))).sum()

# Resultados
print(" Registros válidos en otras variables:\n", valid_counts)
print("\n Valores ausentes:\n", missing_counts)
print("\n Posibles valores mal registrados o erróneos:\n", outliers)
```
